Step 3: I import the request library, which enables me to issue an HTTP GET request to the given URL. Hence, retrieves the HTML data that the server sends back and stores that data in a Python object.

Note: I received a JavaScript code as a response and This code looks completely different from what I saw when I inspected the page with my browser’s developer tools.

```python
!source /content/theanoEnv/bin/activate; 
import requests
URL = "https://realpython.github.io/fake-jobs/jobs/senior-python-developer-0.html"
page = requests.get(URL)
print(page.text)
```

Step 4: To transform the above javascript code, I will proceed to intall the "Beautiful Soup" tool, which is a python library that enables me to parse the javascript code into an HTML structed data.

```python
!source /content/theanoEnv/bin/activate; python -m pip install beautifulsoup4
```

```python
!source /content/theanoEnv/bin/activate;
import requests
from bs4 import BeautifulSoup
URL = "https://realpython.github.io/fake-jobs/"
page = requests.get(URL)
```

```python
!source /content/theanoEnv/bin/activate;
soup = BeautifulSoup(page.content, "html.parser")
```

Step 5: i will proceed to find each element of the already parsed HTML code by ID and By class name using the ".find_all()" and the ".find():" code snippet.
Moreover, I also make use of the ".prettify()" code snippet to make the line of code look neat and pretty.

```python
!source /content/theanoEnv/bin/activate;
results = soup.find(id="ResultsContainer")
```

```python
!source /content/theanoEnv/bin/activate;
print(results.prettify())
```

```python
!source /content/theanoEnv/bin/activate;
job_elements = results.find_all("div", class_="card-content")
for job_element in job_elements:
    print(job_element, end="\n"*2)
```

Step 6: Finally, to Extract the text from each element, i make use of the ".text" code snippet and the ".stip()" code snippet to eliminate ever extra space.


CONCLUSION: The requests library gave me a user-friendly way to fetch static HTML from the Internet using Python. Hence, I can then parse the HTML with another package called Beautiful Soup. Both packages are trusted were helpful companions for this web scraping adventure.

```python
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    print(title_element.text.strip())
    print(company_element.text.strip())
    print(location_element.text.strip())
    print()
```
